```python
df=pd.read_csv("movies.csv")
df1=pd.read_csv("ratings.csv")
df1['timestamp'] = pd.to_datetime(df1['timestamp'],unit='s') # correcting the data type of timestamp from int64 to datetime64[ns]
print(df1)
#6 create three new column name as year, month and quater 
df1['year']=df1['timestamp'].dt.year
df1['month']=df1['timestamp'].dt.month
df1['quater']=df1['timestamp'].dt.quarter
df1
```

```python
#6 find out how many number of movies has been watched in each quater
mov=df1.groupby(['year','quater']).size()
mov
```

```python
#7 what are the average count of user watching movie per year 
movyr=df1.groupby(['year','userId'])
movyr.size()
```

```python
#8 do a group by on movies column and find out the average rating and the total number of users who have watched that movie
# firstly we merge the movie and the ratings data frame for this question 
import pandas as pd

merged_df = pd.merge(df, df1, on='movieId')
merged_df
```
