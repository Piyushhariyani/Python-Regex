```python
import pandas as pd
df=pd.read_feather("Sales_data.ftr")
df
```

```python
df.describe()   # decribing 
```

```python
df.isnull().sum() # checking if there are any null values 
```

```python
df.dropna(inplace=True,ignore_index=True) # dropping all the null va
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.info()
```

```python
# For DD/MM/YYYY
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y', errors='coerce')

# For MM/DD/YYYY
# df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y', errors='coerce')

# For YYYY-MM-DD
# df['Order Date'] = pd.to_datetime(df['Order Date'], format='%Y-%m-%d', errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True,ignore_index=True) # dropping all the null values 
```

```python
df.isnull().sum()
```

```python
df['Month']=df['Order Date'].dt.month
```

```python
df['Month'].unique()
```

```python
df['Month'].dtype
```

```python
df
```

```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'],errors='coerce')
df["Price Each"] = pd.to_numeric(df["Price Each"], errors='coerce')
```

```python
# df[['Year', 'Month', 'Day']] = df['Order Date'].str.split('-', expand=True)
df['Sales']=df['Quantity Ordered'] * df['Price Each']
```

```python
df
```

```python
df['Sales'].sum()    # Total sales
```

```python
# Sales by month 
df.groupby('Month')['Sales'].sum()
```

```python
# Best selling product 
best_product=df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
best_product
```

```python
# Count how many orders per product 
print(df.groupby('Product')['Quantity Ordered'].sum())
```

```python
# Total revenue per product 
print(df.groupby('Product')['Sales'].sum())
```

```python
# highest sales row ('Largest single order')
df[df['Sales']==df['Sales'].max()]
```
